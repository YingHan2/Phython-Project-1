```python
import math
from matplotlib import pyplot as plt

x = list(range(0, 10))
y = [math.pow(num, 2) for num in x]
plt.plot(x, y, color='red')
plt.title('Simple Line Plot')
plt.xlabel('X Values from 0 to 9')
plt.ylabel('Squared Values')
plt.show()
```
